Exercise 1 : Multivariate Normal Distribution

```python
import matplotlib.pyplot as plt
from scipy.stats import norm

univariate_data = norm.rvs(size=1000)

plt.hist(univariate_data, bins=30, density=True, alpha=0.6, color='g')
plt.title('Univariate Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
import numpy as np

mean = [0, 0]
cov = [[1, 0.5], [0.5, 1]]

multivariate_data = multivariate_normal.rvs(mean=mean, cov=cov, size=1000)

plt.scatter(multivariate_data[:, 0], multivariate_data[:, 1], alpha=0.6)
plt.title('Multivariate Normal Distribution')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

Exercise 2: Advanced Probability Distributions


Poisson Distribution:

Description:
The Poisson distribution is used to model the number of events that occur in a fixed interval of time or space, where the events occur with a known average frequency and independently of each other.

Real-world example:
Modeling the number of customers arriving at a store in an hour.

Use:

Suppose that on average 10 customers arrive at a store per hour.
The Poisson distribution can be used to calculate the probability that a certain number of customers (e.g. exactly 12 customers) will arrive in the next hour.
This is useful for forecasting and managing inventory or staffing.
Assumptions:

Events occur independently.
The average number of events () is known and constant.
Events occur one at a time (no double arrivals).

Exercise 3: Statistical Hypothesis Testing


```python
import numpy as np
import pandas as pd
from scipy.stats import f_oneway

np.random.seed(0)

region1 = np.random.normal(20000, 3000, 30)
region2 = np.random.normal(22000, 3500, 30)
region3 = np.random.normal(25000, 5000, 30)

sales_data = pd.DataFrame({
    'Region 1': region1,
    'Region 2': region2,
    'Region 3': region3
})

f_value, p_value = f_oneway(region1, region2, region3)

print('F-value:', f_value)
print('P-value:', p_value)
```

Result:

F-value shows how much the means differ between the groups.
P-value indicates statistical significance. If P-value is less than 0.05, then the differences between the groups are considered statistically significant.
Interpretation:

P-value < 0.05, we reject the null hypothesis and claim that there are significant differences between sales in different regions.

Exercise 4: Linear Regression Analysis


```python
import numpy as np
import pandas as pd
from scipy.stats import linregress

np.random.seed(0)

X = np.random.rand(100) * 50
Y = 2.5 * X + np.random.randn(100) * 10

linear_regression_data = pd.DataFrame({
    'Hours Studied': X,
    'Test Score': Y
})
slope, intercept, r_value, p_value, std_err = linregress(X, Y)

print('Slope:', slope)
print('Intercept:', intercept)
print('R-squared:', r_value**2)
```

Slope shows how the grade changes when studying for one hour more.
Intercept shows the grade a student would get if they didn't study at all.
R-squared shows how well the model explains the variability in the data. A value closer to 1 indicates a strong relationship.
Interpretation:

If the slope is positive, it means that the more hours a student studies, the higher their grade.
If R-squared is close to 1, it means that study time explains test scores well.
